```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a new graph
G = nx.Graph()

# Add nodes (fact table + dimensions)
G.add_node("fact_sales", color='lightblue', size=2000)
dimensions = ["dim_product", "dim_customer", "dim_time", "dim_store"]

# Connect each dimension to the fact table
for dim in dimensions:
    G.add_node(dim, color='lightgreen', size=1500)
    G.add_edge("fact_sales", dim)

# Extract colors and sizes from node attributes
colors = [G.nodes[node]['color'] for node in G.nodes]
sizes = [G.nodes[node]['size'] for node in G.nodes]

# Layout for the star schema
pos = nx.spring_layout(G, center=(0,0), k=1.5)

# Draw the schema
plt.figure(figsize=(8,6))
nx.draw(
    G, pos,
    with_labels=True,
    node_color=colors,
    node_size=sizes,
    font_size=10,
    font_weight='bold',
    edge_color='gray'
)
plt.title("Star Schema Diagram - Retail Data Warehouse", fontsize=14, fontweight='bold')
plt.axis('off')

# Save as PNG
plt.savefig("star_schema.png", dpi=300)
plt.show()
```
